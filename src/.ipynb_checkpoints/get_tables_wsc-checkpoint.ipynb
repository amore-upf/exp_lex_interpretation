```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import spearmanr
%matplotlib inline  
```

```python
all_scores = pd.read_csv('../all_scores.csv', sep= '\t', index_col = 0)
all_scores["model"] = all_scores['mode'] + '-' + all_scores['tied']
all_scores = all_scores[['model', 'component', 'in_vocab']]
all_scores.columns = ['model', 'component', 'corr']
all_scores.set_index('model')
table = all_scores.pivot(index = 'model',columns='component', values='corr')
table = table[sorted(table.columns, reverse = True)]
avg_context_cols = [x for x in table.columns if 'avg' in x]
avg_context = table[avg_context_cols]
table = table.drop(avg_context_cols, axis = 1)

table.columns = ['$' + col + '$'  for col in table.columns]
print(table.to_latex(escape=False))

print(avg_context.to_latex(escape=False))
```

```python
all_scores = pd.read_csv('../results_wsc/all_scores.csv', sep= '\t', index_col = 0)
all_scores["model"] = all_scores['mode'] + '_' + all_scores['tied']
all_scores = all_scores[['model', 'component', 'all_words']]
all_scores.columns = ['model', 'component', 'corr']
all_scores.set_index('model')
table = all_scores.pivot(index = 'model',columns='component', values='corr')
table.columns = ['$' + col + '$'  for col in table.columns]
print(table.to_latex(escape=False))
```
