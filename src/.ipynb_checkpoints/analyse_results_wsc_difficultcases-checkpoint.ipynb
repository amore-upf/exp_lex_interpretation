```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import spearmanr
%matplotlib inline  
```

```python
model_name = '/home/u153390/Dropbox/Research/programming/words-in-context/cws_tied_wiki80_standard_nlayers2_lr0.001_batchsize64_dropout0.2_embsize600_hiddensize600_tied_True.csv'
data = pd.read_csv(model_name, sep = '\t', index_col = 'items')
```

## Easiest for word embeddings

```python
plt.scatter(data['w_t'], data['targets'], color = 'Orange')
```

```python
data['targets_norm'] = data['targets']/10.0
data['diff emb'] = (data['targets_norm'] - data['w_t']).abs()
data['diff sum'] = (data['targets_norm'] - data['w_t + \hat{w_t}']).abs()
data['diff pred'] = (data['targets_norm'] - data['\hat{w_t}']).abs()
```

```python
data_sorted_emb = data.sort_values('diff emb', ascending = True)
```

```python
easiest_emb = data_sorted_emb.head(200)
print('easiest embedding', spearmanr(easiest_emb['targets'], easiest_emb['w_t']))
print('easiest prediction', spearmanr(easiest_emb['targets'], easiest_emb['\hat{w_t}']))
print('easiest sum', spearmanr(easiest_emb['targets'], easiest_emb['w_t + \hat{w_t}']))
print('easiest sum sim', spearmanr(easiest_emb['targets'], easiest_emb['sim- w_t + \hat{w_t}']))
print('easiest sum filter', spearmanr(easiest_emb['targets'], easiest_emb['filter w_t + \hat{w_t}']))

print('\n\n')
print('embedding', spearmanr(data['targets'], data['w_t']))
print('prediction', spearmanr(data['targets'], data['\hat{w_t}']))
print('sum', spearmanr(data['targets'], data['w_t + \hat{w_t}']))
```

```python
plt.scatter(easiest_emb['w_t'], easiest_emb['targets'], color = 'Orange')
plt.scatter(easiest_emb['\hat{w_t}'], easiest_emb['targets'], color = 'Teal')
plt.scatter(easiest_emb['w_t + \hat{w_t}'], easiest_emb['targets'], color = 'Purple')
```

```python
difficult_emb = data_sorted_emb.tail(200)
print('difficult embedding', spearmanr(difficult_emb['targets'], difficult_emb['w_t']))
print('difficult prediction', spearmanr(difficult_emb['targets'], difficult_emb['\hat{w_t}']))
print('difficult sum', spearmanr(difficult_emb['targets'], difficult_emb['w_t + \hat{w_t}']))
print('difficult sum sim', spearmanr(difficult_emb['targets'], difficult_emb['sim- w_t + \hat{w_t}']))
print('difficult sum filter', spearmanr(difficult_emb['targets'], difficult_emb['filter w_t + \hat{w_t}']))

print('\n\n')
print('embedding', spearmanr(data['targets'], data['w_t']))
print('prediction', spearmanr(data['targets'], data['\hat{w_t}']))
print('sum', spearmanr(data['targets'], data['w_t + \hat{w_t}']))
```

```python
plt.scatter(difficult_emb['w_t'], difficult_emb['targets'], color = 'Orange')
plt.scatter(difficult_emb['\hat{w_t}'], difficult_emb['targets'], color = 'Teal')
plt.scatter(difficult_emb['w_t + \hat{w_t}'], difficult_emb['targets'], color = 'Purple')
```

- Where the word embeddings make the most mistakes, the prediction has decent correlation (the only one). 
- Many of these cases are in fact 'same words-different contextual meaning' and 'dissimilar words-same contextual meaning'
- 'same words-different contextual meaning': word embeddings just give 1, sum moves them a bit further away but not enough, the prediction seems a bit better (cannot tell)
- 'dissimilar words-same contextual meaning': word embeddings just assign them low similarity scores, sum is behaving similar, prediction seems a bit better... generally words are rated too dissimilar!
